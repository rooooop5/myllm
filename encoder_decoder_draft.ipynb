```python
import pickle
with open("merge_rules.pkl","rb") as f:
  merge_rules=pickle.load(f)
```

```python
text="i just got a new copy today what are you doing main you why do do this"
text
```

```python
def toUni(ids):
  ids=ids.encode("utf-8")
  ids=list(map(int,ids))
  return ids
```

```python
uni_a=toUni(text);
print(uni_a)
len(uni_a)
```

```python
def _from_uni(ids):
        """Convert list of codepoints back to string"""
        return "".join([chr(i) for i in ids])
```

```python
def encode(text):
  uni_text=toUni(text)
  newids=[]
  while True:
    i=0
    found_merge=0
    while i<len(uni_text)-1:
      tup=(uni_text[i],uni_text[i+1])
      if tup in merge_rules:
        newids.append(merge_rules[tup])
        i=i+2
        found_merge=1
      else:
        newids.append(uni_text[i])
        i=i+1
    if i == len(uni_text) - 1:
      newids.append(uni_text[-1])
    if(found_merge==0):
      break
    else:
      uni_text,newids=newids,[]

  return uni_text
```

```python
rev_merge_rules={}
for k,v in merge_rules.items():
  rev_merge_rules[v]=k
print(rev_merge_rules)
with open("rev_merge_rules.pkl","wb") as f:
  pickle.dump(rev_merge_rules,f)
```

```python
uni_a=encode(text)
print(uni_a)
```

```python
while True:
  i=0
  flag=0
  decoded=[]
  while i<len(uni_a):
    if uni_a[i] in rev_merge_rules:
      decoded.append(rev_merge_rules[uni_a[i]][0])
      decoded.append(rev_merge_rules[uni_a[i]][1])
      flag=1
    else:
      decoded.append(uni_a[i])
    i=i+1
  if(flag==0):
    break
  else:
    uni_a=decoded.copy()



```
