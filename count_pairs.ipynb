```python
with open("dataset.txt","r",encoding="utf-8") as f:
  chars=f.read()
print(len(chars))
```

```python
def toUni(chars):
  ids=chars.encode("utf-8")
  ids=list(map(int,ids))
  return ids
ids=toUni(chars)
print(len(ids))
print(max(ids))
```

```python
def count_pairs(ids):
  count={}
  for pair in zip(ids,ids[1:]):
    count[pair]=count.get(pair,0)+1
  return count

byte_pairs=count_pairs(ids)
```

```python
pairs={}
for key,val in byte_pairs.items():
  pairs[val]=key
top_pair=max(pairs.items())
print(top_pair)
```

```python
newids=[]
i=0
while i<len(ids)-1:
  tup=(ids[i],ids[i+1])
  if tup==top_pair[1]:
    newids.append(227)
    i+=2
  else:
    newids.append(ids[i])
    i+=1
if i==len(ids)-1:
  newids.append(ids[i])
print(len(newids))
```
